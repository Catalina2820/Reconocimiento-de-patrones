```python
#Librerias
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Im√°genes
bosque = cv2.imread('/content/drive/MyDrive/Semestre 8/ReconocimientoDePatrones/RetoSemanaDos/Wild.jpg')
caballo = cv2.imread('/content/drive/MyDrive/Semestre 8/ReconocimientoDePatrones/RetoSemanaDos/Horse.jpg')
rta = cv2.imread('/content/drive/MyDrive/Semestre 8/ReconocimientoDePatrones/RetoSemanaDos/Resultado.jpg')

bosque = bosque[:,:,[2,1,0]]
caballo = caballo[:,:,[2,1,0]]
rta = rta[:,:,[2,1,0]]
#Graficas
plt.figure(figsize=(25,15))
plt.subplot(3,1,1)
plt.imshow(bosque)
plt.title('Bosque')
plt.subplot(3,1,2)
plt.imshow(caballo)
plt.title('Caballo')
plt.subplot(3,1,3)
plt.imshow(rta)
plt.title('Resultado esperado')
```

```python
#Tama√±o del bosque
#Dimensiones para recorrer la imagen grande (bosque)
[Fl, Cl, Ch]=bosque.shape
print('Filas = ',Fl,'\nColumnas = ',Cl, '\nCanales = ',Ch)
```

```python
#Tama√±o del caballo
#y as√≠ definir el tama√±o de la imagen de corte que guarde la relaci√≥n de la original
[Fc,Cc,Cnc] = caballo.shape
print('Filas = ',Fc,'\nColumnas = ',Cc, '\nCanales = ',Cnc)
```

```python
#Haciendo el bosque m√°s grande
bosqueG = cv2.resize(bosque, (Cl*4, Fl*4))
plt.imshow(bosqueG)
```

```python
bosque2 = bosqueG
caballo2 = caballo
```

```python
#Recorriendo la imagen posici√≥n por posici√≥n
for i in range(600,600+Fc,1):
  for j in range(900,900+Cc,1):
    ii = i - 600
    jj = j - 900
    if (caballo2[ii, jj, 0] + caballo2[ii, jj, 1] + caballo2[ii, jj, 2]) <=3:
      caballo2[ii,jj,0] = bosque2[i,j,0]
      caballo2[ii,jj,1] = bosque2[i,j,1]
      caballo2[ii,jj,2] = bosque2[i,j,2]

plt.title('Mi Primera Chamba')
plt.imshow(caballo2)
```
