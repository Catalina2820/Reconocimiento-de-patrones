#3. Operando im치genes:

```python
import cv2 # OpenCV para computer vision
import numpy as np # Para c치lculo de matrices
import matplotlib.pyplot as plt #Para graficar
import os #Habilita el manejo de directorios

#Leyendo una imagen a color y organizando los canales
path = r'/content/drive/MyDrive/Cursos IUPB 2024-1/Reconocimiento de patrones/Material de clase/Semana 2/Procesando ima패genes/Lena.png'
Img_3 = cv2.imread(path)
Img_3 = cv2.cvtColor(Img_3,cv2.COLOR_BGR2RGB)
plt.imshow(Img_3,vmax=255, vmin=0)
plt.show()
#Calculando dimensiones de la imagen
try:
    [Fl, Cl, Ch]=Img_3.shape #Genera dos variables para almacenar el n칰mero de filas y Columnas
except:
    [Fl, Cl]=Img_3.shape #Genera dos variables para almacenar el n칰mero de filas y Columnas
    Ch=1
print('Dimensiones originales: ',Fl,Cl,Ch) #Muestra en pantalla el n칰mero de filas
print('')


#Transformando a grises
Img_3_Gray = cv2.cvtColor(Img_3, cv2.COLOR_BGR2GRAY )
#Visualizando imagen en grises
plt.imshow(Img_3_Gray,cmap='gray',vmax=255, vmin=0) #Grafica la imagen en campo de grises
plt.show()
#Calculando dimensiones de la imagen
try:
    [Fl2, Cl2, Ch2]=Img_3_Gray.shape #Genera dos variables para almacenar el n칰mero de filas y Columnas
except:
    [Fl2, Cl2]=Img_3_Gray.shape #Genera dos variables para almacenar el n칰mero de filas y Columnas
    Ch2=1
print('Dimensiones nuevas: ',Fl2,Cl2,Ch2) #Muestra en pantalla el n칰mero de filas
print('')
```

```python
#Procesando imagen en grises
import numpy as np
Resultado = np.zeros((Fl2,Cl2))
for i in range(Fl2):
  for j in range(Cl2):
    if Img_3_Gray[i,j]>=120:
      Resultado [i,j]= 200
    else:
      Resultado [i,j]= Img_3_Gray[i,j] - 20
plt.imshow(np.uint8(Resultado),cmap='gray',vmax=255, vmin=0) #Grafica la imagen en campo de grises
plt.show()
print('') #Deja espacio vac칤o
```

# 4. Recorte de im치genes:

  En ocasiones es necesario s칩lo concentrarse en una regi칩n de la imagen, e incluso,  guardar esa regi칩n como una imagen nueva.

```python
#Importando librer칤as
import cv2 # OpenCV para computer vision
import numpy as np # Para c치lculo de matrices
import matplotlib.pyplot as plt #Para graficar
import os #Habilita el manejo de directorios

#Leyendo una imagen a color
path = r'/content/drive/MyDrive/Cursos IUPB 2024-1/Reconocimiento de patrones/Material de clase/Semana 2/Procesando ima패genes/Lena.png'
Img_2 = cv2.imread(path)#El orden es BGR
Img_3 = cv2.cvtColor(Img_2,cv2.COLOR_BGR2RGB)
plt.imshow(Img_3,vmax=255, vmin=0)
plt.show()
#Calculando dimensiones de la imagen
try:
    [Fl, Cl, Ch]=Img_3.shape #Genera dos variables para almacenar el n칰mero de filas y Columnas
except:
    [Fl, Cl]=Img_3.shape #Genera dos variables para almacenar el n칰mero de filas y Columnas
    Ch=1
#Mostrando dimensiones en pantalla
print('Dimensiones: ',Fl,Cl,Ch) #Muestra en pantalla el n칰mero de filas
print('') #Deja espacio vac칤o

#####
#Recortando regi칩n de inter칠s de imagen en color
ROI= Img_3[181:320,151:310,:]
#Visualizando imagen recortada
print('Recorte')
plt.imshow(ROI,vmax=255, vmin=0)
plt.show()
try:
    [Fl2, Cl2, Ch2]=ROI.shape #Genera dos variables para almacenar el n칰mero de filas y Columnas
except:
    [Fl2, Cl2]=ROI.shape #Genera dos variables para almacenar el n칰mero de filas y Columnas
    Ch2=1
#Mostrando informaci칩n en pantalla
print('Dimensiones ROI: ',Fl2,Cl2,Ch2) #Muestra en pantalla el n칰mero de filas
print('') #Deja espacio vac칤o

#Guardando resultado
Destino = r'/content/drive/MyDrive/Cursos IUPB 2024-1/Reconocimiento de patrones/Material de clase/Semana 2/Procesando ima패genes/'
Nombre= 'ROI_Lena.bmp'
#La reorganizamos en color
ROI_2 = cv2.cvtColor(ROI,cv2.COLOR_RGB2BGR)
#Guardando
cv2.imwrite((Destino+Nombre), ROI_2)
```
